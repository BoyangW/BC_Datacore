```python
#align the goal key to upload time (all)
def align_goals(result_all, df_goal):
    periodnumList = []
    for rows in range(result_all.shape[0]):
        df_sub = df_goal[df_goal['study_id'] == result_all['study_id'][rows]]
        index = 0
        for i in list(df_sub['goal_start_date']):
            if (result_all['upload_time'][rows] >= i):
                index = index + 1
            else:
                break
        index = index - 1
        periodnum = list(df_sub['period_num'])[index]
        periodnumList.append(periodnum)

    result_all['goal'] = periodnumList
    return result_all

baseline_EDA = align_goals(baseline_EDA, df_goal)
copy_EDA = align_goals(copy_EDA, df_goal)
```

```python
def Stats_List(l):
    Max = np.max(l)
    Min = np.min(l)
    Mean = np.mean(l)
    Median = np.median(l)
    STD = np.std(l)

    #outlier percentage
    q1 = np.percentile(l, 25)  
    q3 = np.percentile(l, 75)
    IQR = q3-q1
    low = q1 - 1.5*IQR
    high = q3+1.5*IQR

    outliersPercent = len([i for i in l if i < low or i > high]) / len(l) * 100

    return([Min, Max, Mean, Median, STD, outliersPercent])
```

```python
def Stats_1(df, column):
    ID_list = list(set(df['ID']))
    
    stats = []
    for id in ID_list:
        df_user = df[df['ID'] == id]

        l = list(df_user[column])
        Max = np.max(l)
        Min = np.min(l)
        Mean = np.mean(l)
        Median = np.median(l)
        STD = np.std(l)

        #outlier percentage
        q1 = np.percentile(l, 25)  
        q3 = np.percentile(l, 75)
        IQR = q3-q1
        low = q1 - 1.5*IQR
        high = q3+1.5*IQR

        outliersPercent = len([i for i in l if i < low or i > high]) / len(l)
        stats.append([id, Min, Max, Mean, Median, STD, outliersPercent])
    return(stats)

def Stats_2(df, column):

    l = list(df[column])
    Max = np.max(l)
    Min = np.min(l)
    Mean = np.mean(l)
    Median = np.median(l)
    STD = np.std(l)

    #outlier percentage
    q1 = np.percentile(l, 25)  
    q3 = np.percentile(l, 75)
    IQR = q3-q1
    low = q1 - 1.5*IQR
    high = q3+1.5*IQR

    outliersPercent = len([i for i in l if i < low or i > high]) / len(l)

    return([Min, Max, Mean, Median, STD, outliersPercent])
```

```python
def shimmer(df_pa_device):
    #get day-level date
    df_pa_device['upload_time'] = df_pa_device['tstamp'].astype('string').str[:10]
    
    #active minute
    df_pa_device_active = df_pa_device[df_pa_device['pa_count'] >= 60]
    
    #group by user and each day, count total physical activity minutes per day
    df_pa_device_count = ps.sqldf("select user_id, upload_time, count(pa_count) as pa_minute_shimmer from df_pa_device_active group by user_id, upload_time")
    
    df_pa_device_count['pa_minute_shimmer'] = df_pa_device_count['pa_minute_shimmer'].fillna(0)
    
    return(df_pa_device_count)
```
